**Определение**. Взвешенный граф назовём бинарным, если каждому его ребру соответствует одна из двух меток: $-1$ или $1$.

**Определение**. Назовём граф $K_3$ сбалансированным сумма весов его ребёр равна $3$ или $-1$.

Будем называть сбалансированные $K_3$ графы $BK_{3,3}$ и $BK_{3,-1}$ соответсвенно.

**Определение**. Полный граф называется сбалансированным, если любой его подграф $BК_3$ сбалансирован.

**Лемма 1**. Сбалансированные графы существуют.

*Доказательство*. Пусть дан граф $K_n = (V, E)$. Покажем, что можно задать веса так, что граф $K_n$ будет $BK_n$. 

Разобьём множество $V = X \sqcup Y$. 

1. Пусть все связи внутри множества $X$ и все связи внутри множества $Y$ положим равными $1$. В таком случае, любой подграф $K_3$ из $X$ или из $Y$ будет сбалансированным ($BK_{3,3}$) в силу того, что сумма весов рёбер будет равна $3$ (т.е. в рамках $X$ и в рамках $Y$ нет рёбер с весом $-1$). 
2. С другой стороны, любая связь между вершинами, принадлежащим различным множествам, будет обозначаться как $-1$. Тогда любой подграф $K_3$, лежащий своими вершинами и в $X$, и в $Y$ будет сбалансированным ($BK_{3,-1}$), так как ровно 2 ребра будут отмечены весом $-1$, что очевидно.

Лемма доказана.

*Замечание 1*: если одно из множеств пусто, тогда все подграфы являются $BK_{3,3}$.
*Замечание 2*: если убрать все рёбра с весом 1, то получим двудольный граф.

**Лемма 2**. Либо все подграфы графа $BK_n$ являются графами $BК_{3,3}$, либо существует не менее $C_{n}^2$ подграфов $K_{3,-1}$. 

*Доказательство*. Если есть хотя бы один подграф $BK_{3,-1}$ графа $BK_n$, то это означает, что оба множества разбиения не пусты согласно *Замечанию 1*. Очевидно, что минимальное число треугольников $BK_{3,-1}$ будет тогда и только тогда, когда либо в $X$, либо в $Y$ есть ровно 1 вершина. Легко посчитать число треугольников, которые будут располагаться своими вершинами в разных множествах. Их количество равно $C_n^2$.

**Замечание**. Число подграфов $BK_{3,-1}$ графа $BK_n$ можно вычислить по формуле $n_xC_{n_y}^2+n_yC_{n_x}^2$, где $n_x, n_y$ -- число вершин во множестве $X$ и $Y$

**Теорема 1**. $K_n$ является графом $BK_{n}$ тогда и только тогда, когда $BK_{n}$ состоит из двух множеств или же $BK_{n}$ состоит только из подграфов $BK_{3,1}$.

*Доказательство*.

*Необходимость*. Необходимость следует из *Замечания 1* и *Леммы 1*.

*Достаточность*. Построим доказательство по индукции. Для случая $BK_{3}$ -- очевидно. Шаг индукции. Пусть дан граф $BK_{n}$. Добавим ещё одну вершину и отнесём её к одному из множеств. Без потери общности, рассмотрим отнесение только к множеству $X$. Тогда соеденим каждую вершину из множества $X$ с новой вершиной ребром с весом $1$. А каждую вершину из множества $Y$ ребром с весом $-1$. Таким образом, оказывается, что разбиение сохранилось. Доказано.

**Теорема 2**. В графе $BK_{n}$ нет ни одного подграфа с ровно двумя положительными рёбрами. Данная теорема  становится очевидной, если принять во внимание построения, проделанные в *лемме 1*.
